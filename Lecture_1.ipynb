<h2>Міри центральної тенденції</h2>


<h3>Мода</h3>
Це значення ознаки, що зустрічається найчастіше. У вибірці може бути кілька чи одна мода.

<h3>Медіана</h3>
Це значення ознаки, яке ділить впорядкову множину порівну. Якщо множина містить парну кількість елементів, то береться середня з двох серединних елементів впорядкованої множини.

<h3>Середнє значення</h3>
Сума всіх значень виміряної ознаки поділяється на кількість виміряних значень.

```python
# Імпортування необхідних бібліотек
import pandas as pd
import numpy as np
from scipy import stats

# Задамо множину
sample = np.array([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])
```

<h6>Використання numpy & scipy</h6>

```python
print(f'Мода: {stats.mode(sample)}\n'
      f'Медіана: {np.median(sample)}\n'
      f'Середнє значення: {np.mean(sample)}')
```

<h6>Використання pandas</h6>

```python
# Задамо множину за допомогою бібліотеки pandas
sample = pd.Series([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])

print(f'Мода: {sample.mode()}\n'
      f'Медіана: {sample.median()}\n'
      f'Середнє значення: {sample.mean()}')
```
