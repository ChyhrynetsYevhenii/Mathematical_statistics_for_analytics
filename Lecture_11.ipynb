<h4>Коваріація</h4>
(Ко - спільна, варіація - мінливість). Міра лінійної залежності двох випадкових величин.

Якщо коваріація позитивна, то зі зростанням значень однієї випадкової величини значення другої мають тенденцію зростати, а якщо знак негативний — то убувати.

де N – кількість випадкових величин, а одиниця – кількість ступенів свободи.

Однак тільки за абсолютним значенням коваріації не можна судити про те, наскільки сильно величини взаємопов'язані, оскільки масштаб коваріації залежить від їх дисперсій. Значення коваріації можна нормувати, поділивши її на твір середньоквадратичних відхилень (квадратного коріння з дисперсій) випадкових величин. Отримана величина називається коефіцієнтом кореляції Пірсона, який завжди знаходиться в інтервалі від -1 до 1:

<h5>Приклад використання кореляції</h5>

```python
'''Демонстрація роботи коваріації та кореляції'''
import numpy as np
import random as r

def cov(x, y):
     assert x.size == y.size
     return ((x - x.mean()) * (y - y.mean())).sum()/(x.size - 1)

def cor(x, y):
     return cov(x, y)/(np.std(x, ddof=1)*np.std(y, ddof=1))

# функція, що імітує випадкові фактори
# р - настільки суттєвим буде випадковий фактор
def randomize(arr, p):
     alpha = np.max(arr) - np.min(arr)
     res = np.zeros(arr.shape)
     for i, v in enumerate(arr):
         sign = 1 if r.choice([True, False]) else -1
         res[i] = v + sign*alpha*r.random()*p
     return res
```

```python
x = np.array(range(30))
y = randomize(x, 0.1)
y1 = randomize(x, 0.5)
y2 = randomize(x, 1)
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 3))
ax1.scatter(x, y)
ax2.scatter(x, y1)
ax3.scatter(x, y2)
ax1.set_title('высокая корреляция')
ax2.set_title('средняя корреляция')
ax3.set_title('низкая корреляция')
plt.show()
```

```python
print(f'''
cov1: {cov(x, y):.2f}
cov2: {cov(x, y1):.2f}
cov3: {cov(x, y2):.2f}

cor1: {cor(x, y):.2f}
cor2: {cor(x, y1):.2f}
cor3: {cor(x, y2):.2f}
''')
```
