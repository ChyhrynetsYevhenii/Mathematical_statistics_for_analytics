<h4>Чому саме квадрат, а не модуль чи куб?</h4>
Можу припустити, що лінійне відхилення більш чутливе до аномалій, квадратичне менше, кубічне — ще менш чутливе. Спробував порахувати для 3-х вибірок: [1,2,3,4,5], [1,2,3,4,50] та [1,2,3,4,500]:

Лінійне: 2.5, 452.5 та 49502.5         
Квадратичне: 1.58, 21.27 та 222.49      
Кубичне: 1.36, 7.68 та 36.71     

Модуль не беруть тому, що модуль – не гладка функція. В нулі у модуля є "злам", через який у похідної відбувається розрив. А дуже багато математичних теорем, які, напевно, будуть потрібні далі, працюють тільки на гладких функціях.

Взагалі, з не гладкими функціями працювати не люблять. Там все стає важче. Тож береться квадрат.

```python
# Імпортування необхідних бібліотек
import pandas as pd
import numpy as np
from scipy import stats

# Задамо множину
sample = np.array([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])
```

```python
# Розрахунок розмаху
print(f'Розмах: {max(sample) - min(sample)}\n'
      f'Розмах за допомогою numpy: {np.ptp(sample)}\n')

# Розрахунок середньоквадратичного відхилення
print(f'Середньоквадратичне відхилення за допомогою numpy: {np.std(sample)}')

sample = pd.Series([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])

print(f'Середньоквадратичне відхилення за допомогою pandas: {sample.std()}')
```

```python
sample = np.array([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])
D = 0
for i in range(len(sample)):
    D += (sample[i] - np.mean(sample))**2/len(sample)
MSE = np.sqrt(D)

print(f'Середньоквадратичне відхилення розраховане без бібліотек: {MSE}')
```
