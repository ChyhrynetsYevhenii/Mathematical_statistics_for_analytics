<h3>Квартилі розподілу</h3><br>
<b>Квартилі</b> - це три точки (значення ознаки), які ділять безліч даних на 4 рівні частини.

<b>Box plot</b> - такий вид діаграми у зручній формі показує медіану, нижній та верхній квартилі, мінімальне та максимальне значення вибірки та аномалії.

Квартілі та inter quartile range використовують, щоб оцінити наявність аномалій. Алгоритм розрахунку - порахували квартилі, порахували різницю між ними, вирахували теоретичний максимум і мінімум, порівняли з наявним і з'ясували чи є у вас аномалії і скільки їх. Якщо багато, то потрібно аналізувати та вирішувати брати їх у вибірку чи ні.

```python
# Імпортування необхідних бібліотек
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Задамо множину
sample = np.array([185, 175, 170, 169, 171, 175, 157, 172, 170, 172, 167, 173, 168, 167, 166,
              167, 169, 172, 177, 178, 165, 161, 179, 159, 164, 178, 172, 170, 173, 171])

# За допомогою діаграми boxplot можемо дізнатись медіану, 2 та 3 квартилі
plt.boxplot(sample, showfliers=1)
plt.show()
```
