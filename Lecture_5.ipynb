<h3>Z-перетворення</h3><br>
Перетворення отриманих даних у стандартну Z-шкалу (Z-scores) із середнім значенням = 0 і дисперсією = 1. Щоб привести до такого виду з кожного спостереження потрібно відібрати середнє значення і розділити на стандартне відхилення.


Іноді необхідно розрахувати z - значення лише окремо взятого спостереження, щоб з'ясувати наскільки далеко воно відхиляється від середнього значення одиницях стандартного відхилення.

<h3>Правило 3ьох-сигм</h3>

```python
''' Вважається, що значення IQ (рівень інтелекту) у людей має нормальний розподіл
із середнім значенням рівним 100 і стандартним відхиленням рівним 15 (M = 100, sd = 15).
Який приблизно відсоток людей має IQ > 125?'''

# Імпортування необхідних бібліотек
from scipy import stats
mean = 100
std = 15
IQ = 125
```

```python
print(f'Тільки у {round(stats.norm(mean, std).sf(IQ)*100, 2)}% населення IQ>{IQ}')
```
