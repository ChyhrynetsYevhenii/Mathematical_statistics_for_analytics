<h3>Довірчі інтервали для середнього</h3>

Якщо ми маємо деяку вибірку та ГС, то ми <b>не можемо точно знати</b> середню ГС, знаючи лише середню вибірку. Однак ми <b>можемо сказати з деяким відсотком впевненості</b>, в якому інтервалі лежить середня ГС. Зрозуміло, що для нас краще, щоб цей інтервал був якнайменше, як це зробити?

Ми знаємо, середня середня вибірка, прагне до середньої ГС, також ми знаємо, що стандартна помилка середнього описує стандартне відхилення розподілу середніх вибірок. Якщо ми візьмемо випадкову вибірку та знайдемо її середню
, а також обчислимо стандартну помилку , ми можемо обчислити довірчий інтевал |X-1.96*se; X+1.96*se|
  який описує середню ГС з деяким інтервалом з 95% довіри.

Загадкове число 1,96 це кількість сигм у нормальному розподілі, необхідні, щоб охопити 95% значень у цьому розподілі.!

<h3>Якщо ми розрахували 95% довірчий інтервал для середнього значення, це означає:</h3>
<ul>
<li><font color='960018'>Середнє значення в генеральній сукупності точно належить розрахованому довірчому інтервалу.</font></li>
<li><font color='green'>Ми можемо бути на 95% впевнені, що середнє значення в генеральній сукупності належить розрахованому довірчому інтервалу.</font></li>
<li><font color='green'>Якщо багаторазово повторювати експеримент, для кожної вибірки розраховувати свій довірчий інтервал, то в 95% випадків справжнє середнє буде всередині довірчого інтервалу.</font></li>
<li><font color='960018'>Середнє значення в генеральній сукупності точно перевищує нижню межу 95% довірчого інтервалу.</font></li>
<li><font color='960018'>Якщо багаторазово повторювати експеримент, то 95 % вибіркових середніх значень будуть належати розрахованому нами довірчому інтервалу. так-так, тут просто треба уявити це в розумі</font></li>
</ul>
<h4>Якщо з лекції засвоїти різницю між середнім ГС та середнім вибірки, а також зрозуміти, що довірчий інтервал будується для вибірки, а не для ГС, то відповіді у тесті легко визначаються.</h4>

```python
'''Обчислення 1.96 за допомогою scipy'''
from scipy import stats

#95%
p = 0.95
# так як у нас двосторонній інтервал, зробимо обчислення
alpha = (1-p)/2
# isf - Inverse survival function (inverse of sf) 
print(f'{stats.norm().isf(alpha):.2f} sigma')
```

```python
'''Розрахуйте 99% довірчий інтервал для наступного прикладу:
середнє = 10, стандартне відхилення = 5, розмір вибірки = 100
'''
import numpy as np
from scipy import stats

p = 0.99
mean = 10
std = 5
n = 100

alpha = (1-p)/2
se = std/np.sqrt(n)
sigma = stats.norm().isf(alpha)
print(f'|{round(mean - se*sigma, 2)}; {round(mean + se*sigma, 2)}|')
```
